```python
from skimage import io
```

```python
img = io.imread('cat2.jpg')
io.imshow(img)
```

```python
from skimage.metrics import structural_similarity as ssim
```

```python
import cv2
```

```python
imgs = {
    'Original': img,
    'Median Filtered': cv2.medianBlur(img, 3),
    'Gaussian Filtered': cv2.GaussianBlur(img, (3, 3), 0),
    'Bilateral Filtered': cv2.bilateralFilter(img, 0, sigmaColor=75, sigmaSpace=75)
}
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 20))
for i, key in enumerate(imgs):
   plt.subplot(2, 4, i + 1)
   plt.title(key)
   # plt.text(-18, -18, f'SSIM: {ssim(imgs[key], img, multichannel=True)}', fontsize=11)
   plt.xlabel(f'SSIM: {ssim(imgs[key], img, multichannel=True)}', fontsize=11)
   plt.imshow(imgs[key])
```
