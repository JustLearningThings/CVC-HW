```python
def RGB2HSV(image):
    for i, _ in enumerate(image):
        for j, _ in enumerate(image[i]):
            r = image[i][j][0] / 255
            g = image[i][j][1] / 255
            b = image[i][j][2] / 255

            cmax = max(r, g, b)
            cmin = min(r, g, b)
            d = cmax - cmin

            h = 0
            if cmax == r:
                h = 60 * (((g - b) / d) % 6)
            elif cmax == g:
                h = 60 * (((b - r) / d) + 2)
            elif cmax == b:
                h = 60 * (((r - g) / d) + 4)

            v = cmax

            if cmax:
                s = d / cmax
            else:
                s = 0
            
            image[i][j][0] = h / 2
            image[i][j][1] = s * 255
            image[i][j][2] = v * 255
    
    return image.astype(int)
```

```python
from skimage import io
import matplotlib.pyplot as plt
```

```python
img = io.imread('dog1.jpg')
img.shape, io.imshow(img)
```

```python
new_img = RGB2HSV(img)
```
