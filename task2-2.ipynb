```python
from skimage import io
import numpy as np
```

```python
img = io.imread('cat1.jpg')
io.imshow(img)
```

```python
def change_contrast_and_brightness(contrast, brightness, image):
    new_img = np.zeros(image.shape, image.dtype)

    for x in range(image.shape[0]):
        for y in range(image.shape[1]):
            for c in range(image.shape[2]):
                new_img[x, y, c] = np.clip(contrast * image[x, y, c] + brightness, 0, 255)
    
    return new_img
```

```python
import matplotlib.pyplot as plt
```

```python
configs = [
    [-10, 10], [-5, 10], [0, 10], [5, 10], [10, 10],
    [-10, 5], [-10, 5], [0, 5], [5, 5], [10, 5],
    [-10, 0], [-5, 0], [0, 0], [5, 0], [10, 0],
    [-10, -5], [-5, -5], [0, -5], [5, -5], [10, -5],
    [-10, -10], [-5, -10], [0, -10], [5, -10], [10, -10]
]
```

```python
for i, conf in enumerate(configs):
    plt.subplot(5, 5, i + 1)
    plt.imshow(change_contrast_and_brightness(conf[1], conf[0], img))
```

```python
io.imshow(change_contrast_and_brightness(-10, 10, img))
```
